```python
import pandas as pd

sales = pd.read_excel("D:\\sales.xlsx")
print(sales.shape)
sales.set_index('Row ID',inplace=True)
sales.head()
```

```python
import matplotlib.pyplot as plt
df=sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', title='Top 10 Best-Selling Products')
plt.show()
```

```python
import matplotlib.pyplot as plt
df=sales
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)
state_sales.head(10).plot(kind='bar', title='Top 10 States by Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
df=sales
segment_sales = df.groupby('Segment')['Sales'].sum()
segment_sales.plot(kind='bar', title='Sales by Customer Segment')
plt.show()
```

```python
import matplotlib.pyplot as plt
df=sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', title='Top 10 Best-Selling Products')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales Trend')
plt.show()
```
