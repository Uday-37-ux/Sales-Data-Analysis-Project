```python
import pandas as pd

sales = pd.read_excel("D:\\sales.xlsx")
print(sales.shape)
sales.set_index('Row ID',inplace=True)
sales.head()
```

```python
sales.info()
```

```python
sales.isnull().sum()
```

```python
# Fill in the Empty Data marked as 'Empty'
sales['Postal Code'].fillna('Empty',inplace=True)
sales.loc[sales['Postal Code'] == 'Empty',['Postal Code']]
```

```python
# Checking again that the Data Structure has no empty fields
sales.info()
```

```python
# Viewing the summary of descriptive statistics from a DataFrame or Series
sales.describe()
```

```python
# Grouping Categories while aggregating Data Frequency
category = sales.groupby(
        ['Category']).agg(
            {'Category':['count']})
        
category.columns = category.columns.droplevel(0)
category = category.rename(columns={'count':'Fr_Category'})

category
```

```python
# Grouping Sub-Categories while aggregating Data Frequency to see more specifically
subCategory = sales.groupby(
        ['Category','Sub-Category']).agg(
            {'Sub-Category':['count']})
        
subCategory.columns = subCategory.columns.droplevel(0)
subCategory = subCategory.rename(columns={'count':'Fr_Sub-Category'})

subCategory
```

```python
# Grouping Segments, Categories, Sub-Categories to perform Frequency and Total Sales aggregation
Segment = sales.groupby(
    ['Segment','Category','Sub-Category']).agg(
        {'Category':['count'],'Sales':['sum']}
    )
Segment.columns = Segment.columns.droplevel(0)

Segment = Segment.rename(columns={'count':'Frequency','sum':'Total Sales'})
Segment
```

```python
# Viewing the Time Difference Between Order and Delivery Based on Shipping Method & Product Category
sales['Times'] = sales['Ship Date'] - sales['Order Date']
sales['Times'] = sales['Times'].astype('str').astype('str')
time = sales.groupby(
    ['Category', 'Ship Mode','Times']
    ).agg(
        {'Times':['count']}
)
time.columns = time.columns.droplevel(0)   
time = time.rename(columns={'count':'Frequency'})

# mengambil Rentang dari Frequency Selisih Waktu di atas 100
time = time.loc[time['Frequency'] >= 100]
time
```

```python
# Analyzing frequently occurring delivery locations
city = sales.groupby(
    ['City']).agg(
        {'State':['count']})
city.columns = ['count']

# Menjumlahkan Frequency State di kelompokkan dengan City
city = city.groupby(['City']).agg({'count': 'sum'})
city.columns = ['sum']
city = city.rename(columns={'sum':'Fr_City'})

# Mengambil Rentang dari Frequency pengiriman di atas 80
# Memprioritaskan Frequency yang besar supaya data lebih valid & akurat dalam suatu Rentang data
city = city.loc[city['Fr_City'] >= 80]
city
```

```python
# Analyzing Customers Who Frequently Purchase Products
customer = sales.groupby(['Region','Customer Name']).agg({'Customer Name':['count']})
customer.columns = customer.columns.droplevel(0)
customer = customer.rename(columns={'count':'Fr_Customer'})

# Mengambil data Frequency dalam rentang lebih dari 15 setiap pembelian
# Memprioritaskan Frequency yang besar supaya data lebih valid & akurat dalam suatu rentang data
customer = customer.loc[(customer.Fr_Customer >= 15)]
customer
```

```python
# Viewing Complete Sales Calculation by Item Category
Category_Sales = sales.groupby(['Category']).agg({'Sales':['min','mean','max','sum']})
Category_Sales.columns = ['min','mean','max','sum']
Category_Sales = Category_Sales.rename(columns={
    'min':'Terendah',
    'mean':'Rata-Rata',
    'max':'Tertinggi',
    'sum':'Total'
    })
Category_Sales
```

```python
# Conducting Sales Calculations Based on Sub-Category of Goods
Sub_Sales = sales.groupby(['Sub-Category']).agg({'Sales':['min','mean','max','sum']})
Sub_Sales.columns = ['min','mean','max','sum']
Sub_Sales = Sub_Sales.rename(columns={
    'min':'Terendah',
    'mean':'Rata-Rata',
    'max':'Tertinggi',
    'sum':'Total'
    })
Sub_Sales
```

```python
# Finding the Standard Deviation
std = sales.groupby(['Category','Sub-Category']).aggregate({'Sales':['std']})
std.columns = std.columns.droplevel(0)
std = std.rename(columns={'std':'Standard Deviation Sales'})

std
```

```python
# Performing Delivery Method Clustering & Aggregating Customer Frequency and Total Sales
method = sales.groupby(['Ship Mode','Segment','Category']).aggregate(
    {'Customer Name':['count'],
     'Sales':['sum']}
    )
method.columns = ['count','sum']
method = method.rename(columns={'count':'Fr_Customer','sum':'Total Sales'})

method
```
